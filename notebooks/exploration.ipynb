# ScribblBot — Data Exploration

```python
import sys
sys.path.insert(0, '..')

import numpy as np
import matplotlib.pyplot as plt
from config import CLASSES, RAW_DIR

cls = 'cat'
data = np.load(RAW_DIR / f'{cls}.npy')
print(f'{cls}: {data.shape}')

fig, axes = plt.subplots(2, 8, figsize=(16, 4))
for i, ax in enumerate(axes.flat):
    ax.imshow(data[i].reshape(28, 28), cmap='gray_r')
    ax.axis('off')
plt.suptitle(cls)
plt.tight_layout()
plt.show()
```
